```julia
using LinearAlgebra
B = [5 -2; -2 5]
# находим собственные значения и векторы
eigen_result = eigen(B)
# извлекаем собственные значения и матрицу собственных векторов
eigenvals = eigen_result.values
eigenvecs = eigen_result.vectors
# вычисляем квадратный корень из собственных значений
sqrt_eigenvals = sqrt.(eigenvals)
# строим диагональную матрицу из квадратных корней собственных значений
D_sqrt = Diagonal(sqrt_eigenvals)
B_sqrt = eigenvecs * D_sqrt * inv(eigenvecs)
B_sqrt
```

```julia
C = [1 -2; -2 1]
eigen_result_C = eigen(C)  # находим собственные значения и векторы
eigenvals_C = eigen_result_C.values  # собственные значения
eigenvecs_C = eigen_result_C.vectors  # матрица собственных векторов
# вычисляем кубический корень из собственных значений
cbrt_eigenvals_C = cbrt.(eigenvals_C)
# строим диагональную матрицу из кубических корней собственных значений
D_cbrt_C = Diagonal(cbrt_eigenvals_C)
# восстанавливаем кубический корень из матрицы:
C_cbrt = eigenvecs_C * D_cbrt_C * inv(eigenvecs_C)
C_cbrt
```

```julia
using LinearAlgebra
A = [140  97  74  168  131;
     97 106  89  131   36;
     74  89 152  144   71;
    168 131 144   54  142;
    131  36  71  142   36]
# Находим собственные значения матрицы A
eigen_result = eigen(A)
eigenvalues = eigen_result.values
println("Собственные значения матрицы A:")
println(eigenvalues)
println()
# Создаём диагональную матрицу из собственных значений
D = Diagonal(eigenvalues)
println("Диагональная матрица D (из собственных значений):")
println(D)
println()
# Создаём нижнедиагональную матрицу из A
LowerTriangular_A = LowerTriangular(A)
# преобразуем в обычную матрицу для вывода (без специальных флагов)
LowerTriangular_matrix = Matrix(LowerTriangular_A)
println("Нижнедиагональная матрица (элементы выше главной диагонали обнулены):")
println(LowerTriangular_matrix)
println()
# Оцениваем эффективность операций
# Время вычисления собственных значений
@time eigen(A);
# Время создания диагональной матрицы
@time Diagonal(eigenvalues);
# Время создания нижнедиагональной матрицы
@time LowerTriangular(A);
println("Эффективность операций:")
println("Вычисление собственных значений (eigen): время выполнения измерено с помощью @time")
println("Создание диагональной матрицы (Diagonal): быстро, так как работает с вектором")
println("Создание нижнедиагональной матрицы (LowerTriangular): эффективно, так как Julia использует оптимизированные операции с треугольными матрицами")
```

```julia
using LinearAlgebra
A_a = [1 2; 3 4]
A_b = (1/2) * [1 2; 3 4]
A_c = (1/10) * [1 2; 3 4]

# Функция для проверки продуктивности матрицы
function is_productive(A)
    n = size(A, 1)
    E = Matrix{Float64}(I, n, n)  # единичная матрица того же размера
    A_diff = E - A
    # Проверяем, существует ли обратная матрица
    if det(A_diff) == 0
        return false, "Матрица (E - A) вырождена, обратная не существует."
    end
    A_inv = inv(A_diff)  # вычисляем (E - A)⁻¹
    # Проверяем, все ли элементы обратной матрицы неотрицательны
    if all(x >= 0 for x in A_inv)
        return true, A_inv
    else
        return false, "Элементы обратной матрицы (E - A)⁻¹ отрицательны."
    end
end

# Проверяем продуктивность каждой матрицы
println("а) Матрица A:")
result_a, info_a = is_productive(A_a)
println("Продуктивна: ", result_a)
if !result_a
    println("Причина: ", info_a)
else
    println("Обратная матрица (E - A)⁻¹: ", info_a)
end
println()

println("б) Матрица (1/2)A:")
result_b, info_b = is_productive(A_b)
println("Продуктивна: ", result_b)
if !result_b
    println("Причина: ", info_b)
else
    println("Обратная матрица (E - A)⁻¹: ", info_b)
end
println()

println("в) Матрица (1/10)A:")
result_c, info_c = is_productive(A_c)
println("Продуктивна: ", result_c)
if !result_c
    println("Причина: ", info_c)
else
    println("Обратная матрица (E - A)⁻¹: ", info_c)
end
```
