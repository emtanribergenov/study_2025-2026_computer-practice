```julia
Z = zeros(Int, 6, 6)
E = ones(Int, 6, 6)

Z1 = copy(Z)
for i in 1:5
    Z1[i, i+1] = 1
    Z1[i+1, i] = 1
end

Z2 = copy(Z)
for i in 1:6
    for j in 1:6
        if (i + j) % 2 == 1
            Z2[i, j] = 1
        end
    end
end

Z3 = copy(Z)
for i in 1:6
    for j in 1:6
        if abs(i - j) == 2
            Z3[i, j] = 1
        end
    end
end

Z3[1,4] = 1; Z3[1,6] = 1
Z3[2,3] = 1; Z3[2,5] = 1
Z3[3,2] = 1; Z3[3,6] = 1
Z3[4,1] = 1; Z3[4,5] = 1
Z3[5,2] = 1
Z3[6,1] = 1; Z3[6,3] = 1

Z4 = copy(Z)
for i in 1:6
    for j in 1:6
        if (i + j) % 2 == 0
            Z4[i, j] = 1
        end
    end
end

println("\nМатрица Z₁:")
println(Z1)
println("\nМатрица Z₂:")
println(Z2)
println("\nМатрица Z₃:")
println(Z3)
println("\nМатрица Z₄:")
println(Z4)
```

```julia
function outer(x, y, operation)
    return operation.(x[:,], y')
end
i = 0:4
j = 0:4
A1 = outer(i, j, (x,y) -> x + y)
x = 0:4
y = [1, 2, 4, 8, 16]
A2 = outer(x, y, (a,b) -> b^a)
i = 0:4
j = 0:4
A3 = outer(i, j, (x,y) -> mod1(x + y, 5) - 1)
n = 10
i = 0:n-1
j = 0:n-1
A4 = outer(i, j, (x,y) -> mod1(x + y, n) .- 1)
i = 0:8
j = 0:8
A5 = outer(i, j, (x,y) -> 8 - abs(x - y))
println("A1:\n", A1)
println("A2:\n", A2)
println("A3:\n", A3)
println("A4:\n", A4)
println("A5:\n", A5)
```

```julia
using Random
Random.seed!(123)
M = rand(1:10, 6, 10)
println("Матрица M:")
println(M)
N = 4
count_greater_than_N = sum(M .> N, dims=2)
println("\nКоличество элементов в каждой строке, больших $N:")
println(vec(count_greater_than_N))

M_value = 7
occurrences = sum(M .== M_value, dims=2)
rows_with_two_occurrences = findall(occurrences .== 2)
println("\nСтроки, где число $M_value встречается ровно 2 раза:")
println(rows_with_two_occurrences)

K = 75
column_pairs = Tuple{Int, Int}[]
for j1 in 1:size(M, 2)
    for j2 in (j1 + 1):size(M, 2)
        if sum(M[:, j1] + M[:, j2]) > K
            push!(column_pairs, (j1, j2))
        end
    end
end

println("\nПары столбцов, сумма элементов которых больше $K:")
println(column_pairs)
```

```julia
sum1 = 0.0
sum2 = 0.0
# Первая сумма: sum(i=1 to 20) sum(j=1 to 5) i^4 / (3 + i*j)
for i in 1:20
    for j in 1:5
        sum1 += i^4 / (3 + j)
    end
end
# Вторая сумма: sum(i=1 to 20) sum(j=1 to 5) i^4 / (3 + j)
for i in 1:20
    for j in 1:5
        sum2 += i^4 / (3 + i * j)
    end
end

println("Первая сумма: ", sum1)
println("Вторая сумма: ", sum2)
```
