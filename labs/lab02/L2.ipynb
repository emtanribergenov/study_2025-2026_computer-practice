```julia
import Pkg; Pkg.add("SpecialFunctions")
using Random
using SpecialFunctions

n = 250
Random.seed!(123)
x = rand(0:999, n)
y = rand(0:999, n)

vector_1 = y[2:end] .- x[1:end-1]
vector_2 = x[1:end-2] .+ 2 .* x[2:end-1] .- x[3:end]
vector_3 = sin.(y[1:end-1]) ./ cos.(x[2:end])
vector_3[abs.(cos.(x[2:end])) .< 1e-10] .= NaN
sum_result = sum(exp.(-x[2:end]) ./ (x[1:end-1] .+ 10))
y_gt_600 = y[y .> 600]
indices_y_gt_600 = findall(y .> 600)
x_corresponding = x[indices_y_gt_600]
x_mean = mean(x)
vector_sqrt_diff = sqrt.(abs.(x .- x_mean))
max_y = maximum(y)
count_within_200 = sum(abs.(y .- max_y) .<= 200)
even_count = sum(iseven.(x))
odd_count = sum(.!(iseven.(x)))
count_multiple_of_7 = sum(x .% 7 .== 0)
x_sorted_by_y = x[sortperm(y)]
top_10_x = sort(x, rev=true)[1:10]
unique_x = unique(x)

println("Вектор 1 (y_i - x_(i-1)), длина: $(length(vector_1))")
println(vector_1)

println("\nВектор 2 (x_i + 2x_(i+1) - x_(i+2)), длина: $(length(vector_2))")
println(vector_2)

println("\nВектор 3 (sin(y_i)/cos(x_(i+1))), длина: $(length(vector_3))")
println(vector_3)

println("\nРезультат суммы: $sum_result")

println("Элементы y > 600: ", y_gt_600)
println("Индексы элементов y > 600: ", indices_y_gt_600)

println("Значения x на тех же позициях: ", x_corresponding)

println("Вектор sqrt(|x – x̄|): ", vector_sqrt_diff)

println("Элементов y, отстоящих от max(y) не более чем на 200: ", count_within_200)

println("Чётных элементов в x: ", even_count)
println("Нечётных элементов в x: ", odd_count)

println("Элементов x, кратных 7: ", count_multiple_of_7)

println("x, отсортированный по возрастанию y: ", x_sorted_by_y)

println("Топ-10 наибольших элементов x: ", top_10_x)

println("Уникальные элементы x: ", unique_x)
```
