```julia
# подключаем необходимые пакеты:
import Pkg
Pkg.add("DifferentialEquations")
using DifferentialEquations
# задаём описание модели с начальными условиями:
a = 0.98
f(u,p,t) = a*u
u0 = 1.0
# задаём интервал времени:
tspan = (0.0,1.0)
# решение:
prob = ODEProblem(f,u0,tspan)
sol = solve(prob)
```

```julia
# подключаем необходимые пакеты:
Pkg.add("Plots")
using Plots
# строим графики:
plot(sol, linewidth=5,title="Модель экспоненциального роста", xaxis="Время",yaxis="u(t)",label="u(t)")
plot!(sol.t, t->1.0*exp(a*t),lw=3,ls=:dash,label="Аналитическое решение")
```

```julia
# задаём точность решения:
sol = solve(prob,abstol=1e-8,reltol=1e-8)
println(sol)
# строим график:
plot(sol, lw=2, color="black", title="Модель экспоненциального роста", xaxis="Время",yaxis="u(t)",label="Численное решение")
plot!(sol.t, t->1.0*exp(a*t),lw=3,ls=:dash,color="red",label="Аналитическое решение")
```

```julia
# подключаем необходимые пакеты:
import Pkg
Pkg.add("DifferentialEquations")
using DifferentialEquations, Plots;
# задаём описание модели:
function lorenz!(du,u,p,t)
σ,ρ,β = p
du[1] = σ*(u[2]-u[1])
du[2] = u[1]*(ρ-u[3]) - u[2]
du[3] = u[1]*u[2] - β*u[3]
end
# задаём начальное условие:
u0 = [1.0,0.0,0.0]
# задаём знанчения параметров:
p = (10,28,8/3)
# задаём интервал времени:
tspan = (0.0,100.0)
# решение:
prob = ODEProblem(lorenz!,u0,tspan,p)
sol = solve(prob)
```

```julia
# подключаем необходимые пакеты:
Pkg.add("Plots")
using Plots
# строим график:
plot(sol, vars=(1,2,3), lw=2, title="Аттрактор Лоренца", xaxis="x",yaxis="y", zaxis="z",legend=false)
```

```julia
# отключаем интерполяцию:
plot(sol,vars=(1,2,3),denseplot=false, lw=1, title="Аттрактор Лоренца", xaxis="x",yaxis="y", zaxis="z",legend=false)
```

```julia
using DifferentialEquations, Plots

# Параметры модели
b = 0.05  # коэффициент рождаемости
c = 0.02  # коэффициент смертности
a = b - c # коэффициент роста популяции
x0 = 100  # начальная численность популяции
tspan = (0.0, 50.0) # интервал времени

# Функция правой части дифференциального уравнения
function maltus_model(dx, x, p, t)
    dx[1] = a * x[1]
end

# Создание задачи и её решение
u0 = [x0]
prob = ODEProblem(maltus_model, u0, tspan)
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)

# Построение графика
plot(sol, linewidth=3, xlabel="Время (t)", ylabel="Численность популяции (x(t))", 
     title="Модель Мальтуса: рост изолированной популяции")
```

```julia
using DifferentialEquations, Plots

# Параметры модели
r = 0.5  # коэффициент роста популяции
k = 500  # предельная численность (ёмкость среды)
x0 = 50  # начальная численность популяции
tspan = (0.0, 50.0) # интервал времени

# Функция правой части дифференциального уравнения
function logistic_model(dx, x, p, t)
    r, k = p
    dx[1] = r * x[1] * (1 - x[1] / k)
end

u0 = [x0]
p = [r, k]
prob = ODEProblem(logistic_model, u0, tspan, p)
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)
plot(sol, linewidth=3, xlabel="Время (t)", ylabel="Численность популяции (x(t))", title="Логистическая модель: рост с огран. ёмкости среды")
```

```julia
using DifferentialEquations, Plots

# Параметры модели
β = 0.3  # коэффициент заражения (интенсивность контактов)
ν = 0.1  # коэффициент выздоровления
N = 1000 # общая численность популяции
S0 = N - 10  # восприимчивые
I0 = 10     # инфицированные
R0 = 0     # переболевшие
u0 = [S0, I0, R0]

tspan = (0.0, 100.0)
# Функция правой части системы дифференциальных уравнений
function sir_model(du, u, p, t)
    β, ν = p
    S, I, R = u
    N = S + I + R
    
    du[1] = -β * S * I / N # изменение числа восприимчивых
    du[2] = β * S * I / N - ν * I # изменение числа инфицированных
    du[3] = ν * I # изменение числа переболевших
end

p = [β, ν]
prob = ODEProblem(sir_model, u0, tspan, p)
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)

plot(sol, linewidth=3, xlabel="Время (t)", 
     ylabel="Численность групп", 
     title="SIR-модель: динамика эпидемии",
     labels=["Восприимчивые (S)" "Инфицированные (I)" "Переболевшие (R)"])
```
